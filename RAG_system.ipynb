```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "BAAI/bge-small-en-v1.5"
)

texts = meta["text"].fillna("").tolist()

print("Total chunks:", len(texts))

embeddings = model.encode(
    texts,
    normalize_embeddings=True,
    show_progress_bar=True
)

print("Embeddings created:", len(embeddings))
```

```python
import faiss
import numpy as np

embeddings = np.array(
    embeddings
).astype("float32")

index = faiss.IndexFlatIP(
    embeddings.shape[1]
)

index.add(embeddings)

print("Indexed:", index.ntotal)
```

```python
def retrieve(question, k=5):

    q_emb = model.encode(
        [question],
        normalize_embeddings=True
    )

    scores, ids = index.search(
        np.array(q_emb).astype("float32"),
        k
    )

    return meta.iloc[ids[0]][
        ["document","section","text"]
    ]
```

```python
import pandas as pd
import numpy as np
import faiss

from sentence_transformers import SentenceTransformer

# Load corpus
meta = pd.read_csv("metaData.csv")

texts = meta["text"].fillna("").tolist()

# Embedding model
model = SentenceTransformer(
    "BAAI/bge-small-en-v1.5"
)

# Create embeddings
embeddings = model.encode(
    texts,
    normalize_embeddings=True
)

embeddings = np.array(
    embeddings
).astype("float32")

# Build FAISS index
index = faiss.IndexFlatIP(
    embeddings.shape[1]
)

index.add(embeddings)

print("RAG Index Ready")

while True:

    question = input("\nQuestion: ")

    if question.lower() == "exit":
        break

    q_emb = model.encode(
        [question],
        normalize_embeddings=True
    )

    scores, ids = index.search(
        np.array(q_emb).astype("float32"),
        5
    )

    print("\nTop Retrieved Chunks\n")

    for idx in ids[0]:

        row = meta.iloc[idx]

        print("=" * 80)

        print("Document:", row["document"])

        print("Section:", row["section"])

        print(row["text"][:500])
```

```python
import pandas as pd

test = pd.read_csv("test.csv")

def answer_question(row):
    return row["contexts"][:1000]

answers = []

for _, row in test.iterrows():
    answers.append(answer_question(row))

submission = pd.DataFrame({
    "id": test["id"],
    "answer": answers
})

submission.to_csv("submission.csv", index=False)

print("Done")
```
